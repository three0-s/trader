```python
import ccxt 
import pprint 
import pandas as pd


binance = ccxt.binance()
# markets = binance.load_markets()
# print(markets.keys())
# print(len(markets))
btc_ohlcv = binance.fetch_ohlcv("BTC/USDT")

print(btc_ohlcv)
```

```python
df = pd.DataFrame(btc_ohlcv, columns=['datetime', 'count', 'open', 'high', 'low', 'close', 'volume', 'VWAP'])
df['Timestamp'] = pd.to_datetime(df['datetime'], unit='ms')
del df['datetime']
df.set_index('Timestamp', inplace=True)

print(df)
```

```python
pair = 'ETH'
df = pd.read_csv(f"/mnt/won/230514{pair}.csv")
```

```python
df['Timestamp'] = (df['Timestamp']/1000).astype('int64')
df['Count'] = df['Volume']*2
df['VWAP'] = (df['Open'] + df['High'] + df['Low'] + df['Close'])/4
```

```python
df.head()
```

```python
cols = df.columns.to_list()
df = df[['Timestamp', 'Count']+cols[1:6]+['VWAP']]
df.head()
```

```python
df.to_csv(f"230514_{pair}_processed.csv", index=False)
```
