```python
from matplotlib import pyplot as plt
import mpl_finance
import pandas as pd


cur = 20
current_map_df = pd.read_csv("/Users/yewon/Documents/traderWon/envs/data/1/1882843.csv")
CPS = [(200, 4),
        (0, 0),
        (230, 10),
        (250, 1)]
position = ['Long 1x', 'Long 2x', 'Short 1x', 'Short 2x']

st = cur-15 if cur >= 15  else 0
pre = current_map_df.iloc[st:cur]
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(1, 2, 1)
ax.set_ylabel("USD")
plt.grid()
ay = fig.add_subplot(1, 2, 2)
ay.axis('off')

prev = -1
for i in range(4):
    if CPS[i][1] != 0:
        ay.text(0, 0.1+(prev+1)/20, f"{position[i]}] CPS: {CPS[i][0]:.2f}$      {CPS[i][1]:d}")
        prev += 1
ay.text(0, 0.7, f"Net Profit Rate: {120/100:.2f}%", color='g')
mpl_finance.candlestick2_ohlc(ax, pre['Open'], pre['High'], pre['Low'], pre['Close'], width=0.5, colorup='r', colordown='b')

plt.show()
```
